```python
import pprint
import json
import uuid

import random
random.seed(17)

import numpy as np
np.random.seed(17) # fix NumPy random number generator seed

import networkx as nx
import matplotlib.pyplot as plt


import os

# add some relative paths so we can import sampling.py module.
import sys
sys.path.append("../")
sys.path.append("../src")
sys_path = sys.path

import sampling
```

```python

relative_path = "../benchmark_instances"
benchmark_instances = os.listdir(relative_path)

G_list = []
G_name_list = []
filename_list = []
print("displaying:")
for filename in benchmark_instances:
    if ".json" in filename:
        print(filename)
        G = sampling.load_graph_instance_from_file(relative_path + "/" + filename)
        G_list.append(G)
        G_name_list.append(G.name)
        filename_list.append(filename)

combined_list = list(zip(G_name_list, G_list, filename_list))
combined_list.sort(key=lambda x: x[0])

for instance in combined_list:
    print("\n\n\n------------------------------------------------------")
    G = instance[1]
    print("Graph Name:", G.name)
    print("filename:", instance[2])
    G.plot()
    adj_matrix, external_field_B = G.get_adjacency_matrix_with_zero_diagonal_as_numpy_matrix()
    plt.imshow(adj_matrix)
    plt.xlabel("node index $j$")
    plt.ylabel("node index $i$")
    plt.title("Adjacency Matrix (zero diagonal)")
    plt.show()
    pprint.pprint(G.get_summary_details())


```
