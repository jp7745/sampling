Calculating the probability distribution of system states directly is not possible for large systems.  We show an example on a small system of $n=9$ nodes.

```python
filename = "../benchmark_instances/instance.a8be0d9a-3238-11ee-b3fb-9309595bcf35.EXAMPLE.json"

G = sampling.load_graph_instance_from_file(filename)

G.plot()
pprint.pprint(G.get_summary_details())
```

```python
# WARNING!  Better keep number of nodes n low!


n = len(G.nodes())
N = 2**n

G.brute_force_probability_distribution_calculation()

plt.stem(G._exact_probability_distribution)
plt.xlabel("State (index)")
plt.ylabel("Probability")
plt.title("Exact Probability Distribution Function")
plt.show()

```

Note that the graph in question has a consistent external field of 0.1 applied to it.  So the extreme points of all spin +1 and all spin -1 are not equal.
